```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train / 255.0
x_test = x_test / 255.0

model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train, y_train, epochs=5)

loss, accuracy = model.evaluate(x_test, y_test)

predictions = model.predict(x_test)

print("\n====================================")
print(" HANDWRITTEN DIGIT RECOGNITION")
print("====================================")
print("Total Training Images :", len(x_train))
print("Total Testing Images  :", len(x_test))
print("Model Accuracy        :", round(accuracy * 100, 2), "%")
print("====================================")

index = 0

predicted_digit = predictions[index].argmax()
actual_digit = y_test[index]

print("\nSample Prediction:")
print("Actual Digit    :", actual_digit)
print("Predicted Digit :", predicted_digit)

plt.imshow(x_test[index], cmap='gray')
plt.title(
    f"Actual: {actual_digit} | Predicted: {predicted_digit}"
)
plt.axis("off")
plt.show()
```
